```python
from etl.data_processor import DataProcessor

def main():
    """
    Main function to run the ETL process for the FPoly survey data.
    """
    input_path = r'../data/raw/fpoly_survey.csv'
    output_path = r'../data/processed/fpoly_survey_cleaned.csv'

    # Initialize and run the processor
    processor = DataProcessor(file_path=input_path)
    cleaned_data_head = processor.process(output_path=output_path)

    print("\n--- ETL Process Complete ---")
    print("Cleaned data saved to:", output_path)
    print("\nFirst 5 rows of the cleaned data:")
    print(cleaned_data_head)

if __name__ == "__main__":
    main()
```
